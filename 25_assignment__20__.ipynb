```python
def filter_integers_only(lst):
    return [item for item in lst if isinstance(item, int)]

# Example usage:
print(filter_integers_only([1, "apple", 42, "banana", 7]))  
# Output: [1, 42, 7]
```

```python
def add_index_to_elements(lst):
    return [num + i for i, num in enumerate(lst)]

# Example usage:
print(add_index_to_elements([1, 2, 3, 4, 5]))   # Output: [1, 3, 5, 7, 9]
print(add_index_to_elements([10, 20, 30]))      # Output: [10, 21, 32]
```

```python
def find_missing_number(lst):
    return sum(range(1, 11)) - sum(lst)

# Example usage:
print(find_missing_number([1, 2, 3, 4, 5, 6, 7, 8, 10]))  # Output: 9
print(find_missing_number([2, 3, 4, 5, 6, 7, 8, 9, 10]))  # Output: 1
```

```python
def missing_num(lst):
    n = len(lst) + 1
    expect = (n * (max(lst) + min (lst))) // 2
    ectual = sum(lst)
    missing = expect - ectual

    return missing

user_input = [1,2,3,4,6,7,8,9,10]
result = missing_num(user_input)
print(result)
```
