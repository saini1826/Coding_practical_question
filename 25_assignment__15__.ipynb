```python
def divisible_by_5_and_7(n):
    """
    Generator that yields numbers divisible by both 5 and 7 from 0 to n.
    """
    for i in range(n + 1):
        if i % 5 == 0 and i % 7 == 0:
            yield i

# Input from user
try:
    n = int(input("Enter the upper limit (n): "))
    result = [str(num) for num in divisible_by_5_and_7(n)]
    print("Numbers divisible by 5 and 7:", ",".join(result))
except ValueError:
    print("Invalid input. Please enter a valid integer.")
```

```python
def even_numbers(n):
    """
    Generator that yields even numbers from 0 to n.
    """
    for i in range(n + 1):
        if i % 2 == 0:
            yield i

# Input from user
try:
    n = int(input("Enter the upper limit (n): "))
    result = [str(num) for num in even_numbers(n)]
    print("Even numbers between 0 and", n, ":", ",".join(result))
except ValueError:
    print("Invalid input. Please enter a valid integer.")
```

```python
# Input from user
try:
    n = int(input("Enter the number of Fibonacci terms (n): "))

    # Generate Fibonacci sequence using list comprehension
    fib = [0, 1]
    [fib.append(fib[-1] + fib[-2]) for _ in range(2, n)]

    # Print comma-separated result
    print("Fibonacci sequence:", ",".join(map(str, fib[:n])))

except ValueError:
    print("Invalid input. Please enter a valid integer.")
```

```python
def extract_username(email):
    """
    Extracts the username from an email address of the form 'username@companyname.com'.
    """
    if '@' in email:
        return email.split('@')[0]
    else:
        return "Invalid email format"

# Input from user
email_input = input("Enter an email address: ")
username = extract_username(email_input)

# Output result
print("Username:", username)
```

```python
class shape:
    def square(self):
        return 0

class square(shape):
    def __init__(self, length):
        self.length = length

    def area(self):
        return self.length * self.length

a = square(5)
print("square is :",a.area())
```
