```python
def count_equal_values(a, b, c):
    
    if a == b == c:
        return 3
    elif a == b or b == c or a == c:
        return 2
    else:
        return 0

print(count_equal_values(1, 1, 1))  # Output: 3
print(count_equal_values(2, 2, 3))  # Output: 2
print(count_equal_values(4, 5, 6))  # Output: 0
```

```python
def dict_to_tuples(d):
   
    return list(d.items())

sample_dict = {'name': 'Alice', 'age': 30, 'city': 'Delhi'}
print(dict_to_tuples(sample_dict))
```

```python
def list_to_dict(data):
    return dict(data)

user_input = [('name', 'Alice'), ('age', 30), ('city', 'Delhi')]
result = list_to_dict(user_input)
print(result)
```

```python
def transform_by_ascii_parity(text):
    result = ''
    for char in text:
        ascii_code = ord(char)
        if char.isalpha():
            if ascii_code % 2 == 0:
                result += char.upper()
            else:
                result += char.lower()
        else:
            result += char  # Keep non-alphabetic characters unchanged
    return result

# Example usage:
print(transform_by_ascii_parity("Hello World!"))  # Output: hEllO wOrlD!
```
