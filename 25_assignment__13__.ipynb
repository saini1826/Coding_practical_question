```python
import math

C = 50
H = 30

d_item = input("Enter comma-separated values of D: ").split(",")

# Calculate Q for each D
results = []
for i in d_item:
        D = int(i.strip())
        Q = math.sqrt((2 * C * D) / H)
        results.append(round(Q, 2))  

print("Calculated Q values:", results)
```

```python
def sorted_comma_saprated(sentence):
    words = sorted(sentence.split(","))
    result = ",".join(words)

    return result

user_input = input("enter a sentence :")
result = sorted_comma_saprated(user_input)
print("sorted word :", result)
```

```python
def count_letters_digits(sentence):
    letters = digits = 0
    for char in sentence:
        if char.isalpha():
            letters += 1
        elif char.isdigit():
            digits += 1
    return letters, digits

# Input from user
text = input("Enter a sentence: ")

# Count and display
letters, digits = count_letters_digits(text)
print(f"Letters: {letters}")
print(f"Digits: {digits}")
```

```python
def generate_2d_array(x, y):
    result = []
    for i in range(x):
        row = []
        for j in range(y):
            row.append(i * j)
        result.append(row)
    return result

try:
    x, y = map(int, input("Enter two digits X,Y separated by comma: ").split(','))
    array = generate_2d_array(x, y)
    
    print("Generated 2D Array:")
    for row in array:
        print(row)
except ValueError:
    print("Invalid input. Please enter two integers separated by a comma.")
```

```python
def generate_list(x, y):
    result = []
    for i in range(x):
        raw = []
        for j in range(y):
            raw.append(i * j)
            result.append(raw)

    return result
```

```python
def duplicate_word(sentence):
    words = sentence.split()
    unique = sorted(set(words))

    return " ".join(unique)

text = input("enter a senence :")
result = duplicate_word(text)
print(result)
```

```python
import re

def is_valid_password(password):

    if len(password) < 6 or len(password) > 12:
        return False
    if not re.search(r"[a-z]", password):
        return False
    if not re.search(r"[A-Z]", password):
        return False
    if not re.search(r"[0-9]", password):
        return False
    if not re.search(r"[$#@]", password):
        return False
    return True

# Input from user
passwords = input("Enter comma-separated passwords to validate: ").split(',')

# Check each password
valid_passwords = [i for i in passwords if is_valid_password(i.strip())]

# Output valid ones
print("Valid passwords:", ','.join(valid_passwords))
```
