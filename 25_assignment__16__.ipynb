```python
def sutter(words):
    if len(words) < 2:
        return f"{words} ... {words}"
    prefix = words[:2]
    return f"{prefix}... {prefix}... {words}.."

user = input("enter a string :")
result = sutter(user)
print(result)
```

```python
import math

def radians_to_degrees(radian):
    degrees = math.degrees(radian)
    return round(degrees, 1)

# Example usage:
print(radians_to_degrees(math.pi))      
print(radians_to_degrees(math.pi / 4))   
print(radians_to_degrees(1))        
```

```python
def is_curzon(num):
    term = 1 + 2 ** num
    _term = 1 + 2 * num
    return term % _term == 0

# Example usage:
print(is_curzon(5))   
print(is_curzon(10))  
print(is_curzon(14))  
```

```python
def decimal_to_binary(n):
    binary = ""

    while n > 0:
        binary = str(n % 2) + binary
        n //=2

    return binary

user_input = int(input("enter a number :"))
result = decimal_to_binary(user_input)
print("binary is :",result)
```
