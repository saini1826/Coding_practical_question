```python
def binary(num):
    binary = ""
    # if num == 0:
    #     return "0"
    while num > 0:
        binary = str(num % 2) + binary
        num //= 2

    return binary

user_input = int(input("enter a number :"))
result = binary(user_input)
print("decimal to binary :",result)
```

```python
def decimal_to_octal(n):
    octal = ""
    
    while n > 0:
        octal = str(n % 8) + octal
        n //= 8
    return octal

num = int(input("Enter a decimal number: "))
result = decimal_to_octal(num)
print(f"The octal representation of {num} is {result}.")
```

```python
def decimal_to_hexadecimal(n):
    hex_digits = "0123456789ABCDEF"
    hexadecimal = ""
    if n == 0:
        return "0"
    while n > 0:
        remainder = n % 16
        hexadecimal = hex_digits[remainder] + hexadecimal
        n //= 16
    return hexadecimal


num = int(input("Enter a decimal number: "))
result = decimal_to_hexadecimal(num)
print(f"The hexadecimal representation of {num} is : {result}.")
```

```python
char = input("Enter a character: ")

if len(char) != 1:
    print("Please enter only one character.")
else:
    ascii_value = ord(char)
    print(f"The ASCII value of '{char}' is {ascii_value}.")
```

```python
def get_ascii_value(char):
     if len(char) != 1:
         return "Please enter a single character."
     else:
         return ord(char)

character = input("Enter a character: ")
result = get_ascii_value(character)

print(f"The ASCII value of '{character}' is {result}.")
```
