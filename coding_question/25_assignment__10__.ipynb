```python
def multiply_list(numbers):
    result = 1
    for num in numbers:
        result = num * result
    return result

my_list = [2, 3, 4, 5]
product = multiply_list(my_list)

print("List:", my_list)
print("Product of all numbers:", product)
```

```python
def find_n_largest(lst, n):
    return sorted(lst, reverse=True)[:n]

number = [12, 4, 24, 6, 8, 10, 7]
n = 3

print(find_n_largest(number, n))
```

```python
list1 = [1, [], 2, [], 3, 4, [], 5]

# using list comprehension
cleaned_list = [item for item in list1 if item != []]

print("Original list:", list1)
print("List after removing empty lists:", cleaned_list)
```

```python
list1 = [1, [], 2, [], 3, 4, [], 5, 12]

lst = []

for item in list1:
    if item != []:
        lst.append(item)

print(lst)
```

```python
original_list = [10, 20, 30, 40]

# Method 1: Using slicing
copy1 = original_list[:]

# Method 2: Using list()
copy2 = list(original_list)

# Method 3: Using copy()
copy3 = original_list.copy()

# Display results
print("Original List:", original_list)
print("Copy using slicing:", copy1)
print("Copy using list():", copy2)
print("Copy using copy():", copy3)
```

```python
my_list = [1, 2, 3, 2, 4, 2, 5]
element = 2
count = 0

for item in my_list:
    if item == element:
        count += 1

print(f"The element {element} appears {count} times.")
```
