```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)

factorial(5)
```

```python
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)


user_input = int(input("enter a number :"))
for item in range(user_input):
    print(fibonacci(item))
```

```python
import math

num = float(input("Enter a positive number: "))


if num <= 0:
    print("Natural logarithm is only defined for positive numbers.")
else:
    ln_value = math.log(num)
    print(f"The natural logarithm of {num} is {ln_value:.4f}")
```

```python
n = int(input("Enter a natural number: "))

cube_sum = ((n * (n + 1)) // 2) ** 2

print(f"The sum of cubes of the first {n} natural numbers is {cube_sum}.")
```
