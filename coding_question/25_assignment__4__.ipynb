```python
def armstrong_num(num):
    num_str = str(num)
    num_len = len(num_str)
    total = sum(int(item) ** num_len for item in num_str)

    return total == num

user_input = int(input("enter a number :"))
result = armstrong_num(user_input)
print(f"this is armstrong number : {result}")
```

```python
def sum_count(num):
    digit = str(num)
    count = 0

    for item in digit:
        count = int(item) + count

    return count

user_input = input("enter a digit :")
result = sum_count(user_input)
print(f"sum of natural number : {result}")
```

```python
start = int(input("Enter the starting number: "))
end = int(input("Enter the ending number: "))

print(f"Armstrong numbers between {start} and {end} are:")

for num in range(start, end + 1):
    num_digits = len(str(num))
    sum_of_powers = 0
    temp = num

    while temp > 0:
        digit = temp % 10
        sum_of_powers += digit ** num_digits
        temp //= 10

    if sum_of_powers == num:
        print(num, end=" ")
```
