```python
def add_matrix(matrix1, matrix2):
    result = []
    for i in range(len(matrix1)):
        raw = []
        for j in range(len(matrix1[0])):
            raw.append(matrix1[i][j] + matrix2[i][j])
        result.append(raw)

    return result

a = [[1, 2, 3],
     [4, 5, 6]]

b = [[7, 8, 9],
     [1, 2, 3]]

add = add_matrix(a, b)
print("Sum of matrices:")
for i in add:
    print(i)
```

```python
def multiply_matrices(A, B):
    result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]

    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    
    return result

A = [[1, 2],
     [3, 4]]

B = [[5, 6],
     [7, 8]]

product = multiply_matrices(A, B)

print("Product of matrices:")
for i in product:
    print(i)
```

```python
def transpose_matrix(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    trans = []

    for j in range(cols):
        row = []
        for i in range(rows):
            row.append(matrix[i][j])
        trans.append(row)
    
    return trans

matrix = [[1, 2, 3],
          [4, 5, 6]]

result = transpose_matrix(matrix)
print("Transposed matrix:")
for i in result:
    print(i)
```

```python
def sort_word(word):
    words = word.split()
    words.sort()
    return words

input_user = "banana apple mango cherry"
result = sort_word(input_user)
print("sort word :", result)
```

```python
def remove_puncuation(sentence):
    remove = str.maketrans("", "", string.punctuation)
    return sentence.translate(remove)

user_input = "Hello, world! Let's, clean' this:'' text."
result = remove_punctuation(user_input)
print("Remove punctuation is :", result)
```
