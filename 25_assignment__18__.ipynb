```python
def filter_integers(lst):
    return [item for item in lst if isinstance(item, int)]

# Example usage:
print(filter_integers([1, "hello", 5, "42", 0, "world", 99]))  
```

```python
def reverser(text):
    return text[::-1].swapcase()

# Example usage:
print(reverser("Hello World"))  # Output: DLROw OLLEh
print(reverser("Python3"))      # Output: 3NOHTYp
print(reverser("aBcD"))         # Output: dCbA
```

```python
# Example list
items = [10, 20, 30, 40, 50]

# Destructuring assignment
first, *middle, last = items

print("First:", first)     # Output: 10
print("Middle:", middle)   # Output: [20, 30, 40]
print("Last:", last)       # Output: 50
```

```python
def move_type_to_end(lst, data_type):
    front = [x for x in lst if not isinstance(x, data_type)]
    back = [x for x in lst if isinstance(x, data_type)]
    return front + back

# Example usage:
print(move_type_to_end([1, 'a', 2, 'b', 3, 'c'], str))
# Output: [1, 2, 3, 'a', 'b', 'c']

print(move_type_to_end([True, 1, False, 2, 3.5, True], bool))
# Output: [1, 2, 3.5, True, False, True]
```

```python
def move_to_end(lst, val):
    front = [x for x in lst if x != val]
    back = [x for x in lst if x == val]
    return front + back

# Example usage:
print(move_to_end([1, 3, 2, 4, 4, 1], 1))        # ➞ [3, 2, 4, 4, 1, 1]
print(move_to_end([7, 8, 9, 1, 2, 3, 4], 9))     # ➞ [7, 8, 1, 2, 3, 4, 9]
print(move_to_end(["a", "a", "a", "b"], "a"))    # ➞ ['b', 'a', 'a', 'a']
```
