```python
def find_largest_digit(list1):
    largest = list1[0]

    for item in list1:
        if item > largest:
            largest = item

    return largest

user_input = [2,3,4,5,6,7,8,9,10]
result = find_largest_digit(user_input)
print("largest digit in array :", result)
```

```python
def rotate_list(lst, k, direction='left'):
    k = k % len(lst)  

    if direction == 'left':
        return lst[k:] + lst[:k]
    elif direction == 'right':
        return lst[-k:] + lst[:-k]
    else:
        raise ValueError("Direction must be 'left' or 'right'")

my_list = [1, 2, 3, 4, 5]
k = 3

print("Left rotation:", rotate_list(my_list, k, 'left'))   
print("Right rotation:", rotate_list(my_list, k, 'right')) 
```

```python
def split_and_append(arr, k):
    
    return arr[k:] + arr[:k]

# Example usage
array = [1, 2, 3, 4, 5, 6]
k = 2

result = split_and_append(array, k)
print("Original array:", array)
print(f"After splitting at index {k} and appending:", result)
```
