```python
def sum_divisible_by_c(a, b, c):
    return sum(x for x in range(a, b + 1) if x % c == 0)

# Example usage:
print(sum_divisible_by_c(1, 10, 2))  # Output: 30 (2 + 4 + 6 + 8 + 10)
print(sum_divisible_by_c(5, 15, 3))  # Output: 45 (6 + 9 + 12 + 15)
print(sum_divisible_by_c(10, 20, 5)) # Output: 75 (10 + 15 + 20)
```

```python
def is_inequality_correct(expression):
    try:
        return eval(expression)
    except:
        return False

# Example usage:
print(is_inequality_correct("3 < 5 < 7"))     # Output: True
print(is_inequality_correct("10 > 20"))       # Output: False
print(is_inequality_correct("5 == 5 != 3"))   # Output: True
print(is_inequality_correct("4 <= 4 < 10"))   # Output: True
```

```python
def replace_vowels(text, replacement_char):
    vowels = "aeiouAEIOU"
    return ''.join(replacement_char if char in vowels else char for char in text)

# Example usage:
print(replace_vowels("Hello World", "*"))   # Output: H*ll* W*rld
print(replace_vowels("Data Science", "#"))  # Output: D#t# Sc##nc#
```

```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)

user_input = int(input("enter a number ;"))
result = factorial(user_input)
print("factorial is :",result)
```

```python
def hamming_distance(str1, str2):
    return sum(c1 != c2 for c1, c2 in zip(str1, str2))

# Example usage:
print(hamming_distance("abcbba", "abcbda"))  # Output: 1
print(hamming_distance("101010", "100100"))  # Output: 2
print(hamming_distance("hello", "hullo"))    # Output: 1
```
