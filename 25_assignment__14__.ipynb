```python
class divisible_by:
    def __init__(self, n):
        self.n = n

    def generator(self):
        for item in range(self.n + 1):
            if item % 7 == 0:
                yield item

if __name__ == "__main__":
    try:
        user_input = int(input("enter a number :"))
        result = divisible_by(user_input)
        for i in result.generator():
            print(i)
    except ValueError as ve:
        print(ve)
    except Exception as e:
        print(e)
    finally:
        print("program is finised")
```

```python
def compute_word_frequency(text):
    words = text.split()
    freq = {}

    for word in words:
        freq[word] = freq.get(word, 0) + 1

    
    sorted_freq = dict(sorted(freq.items()))
    return sorted_freq

# Input from user
input_text = input("Enter a sentence or sequence of words: ")

# Compute and display
result = compute_word_frequency(input_text)
print("Word Frequencies (sorted):")
for word, count in result.items():
    print(f"{word}: {count}")
```

```python
def sorted_words(data):
    words = data.split()
    freq = {}

    for item in words:
        freq[item] = freq.get(item, 0) + 1

    return freq

text = input("enter a sentence :")
result = sorted_words(text)

for word, count in result.items():
    print(f"{word} : {count}")
```

```python
class Person:
    def getGender(self):
        return "Unknown"

class Male(Person):
    def getGender(self):
        return "Male"

class Female(Person):
    def getGender(self):
        return "Female"

# Example usage
m = Male()
f = Female()

print("Male says:", m.getGender())
print("Female says:", f.getGender())
```

```python
subjects = ["I", "You"]
verbs = ["Play", "Love"]
objects = ["Hockey", "Football"]

# Generate all combinations
sentences = []
for sub in subjects:
    for verb in verbs:
        for obj in objects:
            sentences.append(f"{sub} {verb} {obj}")

# Output the sentences
print("Generated Sentences:")
for sentence in sentences:
    print(sentence)
```

```python
def binary_search(sorted_list, target):

    left, right = 0, len(sorted_list) - 1

    while left <= right:
        mid = (left + right) // 2
        if sorted_list[mid] == target:
            return mid
        elif sorted_list[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1

# Example usage
data = [2, 4, 6, 8, 10, 12, 14]
item = 12
index = binary_search(data, item)

if index != -1:
    print(f"Item {item} found at index {index}")
else:
    print(f"Item {item} not found in the list")
```
