```python
def double_characters(text):
    return ''.join(char * 2 for char in text)

# Example usage:
print(double_characters("hello"))     # Output: hheelllloo
```

```python
def reverse_boolean(value):
    if isinstance(value, bool):
        return not value
    else:
        return "boolean expected"

# Example usage:
print(reverse_boolean(True))     # Output: False
print(reverse_boolean(False))    # Output: True
print(reverse_boolean("True"))   # Output: boolean expected
print(reverse_boolean(1))        # Output: boolean expected
```

```python
def capital_indices(s):
    return [i for i, char in enumerate(s) if char.isupper()]

# Example usage:
print(capital_indices("HeLLo WorLD"))  # Output: [0, 2, 3, 8, 9]
print(capital_indices("python"))       # Output: []
print(capital_indices("DataScience"))  # Output: [0, 4]
```

```python
def find_evens(n):
    return [x for x in range(1, n + 1) if x % 2 == 0]

# Example usage:
print(find_evens(10))  # Output: [2, 4, 6, 8, 10]
print(find_evens(7))   # Output: [2, 4, 6]
```
