```python
def divisible_in_range(x, y, n):

    if n == 0:
        raise ValueError("Divisor 'n' cannot be zero.")
    
    return [i for i in range(x, y + 1) if i % n == 0]

print(divisible_in_range(10, 30, 5))
```

```python
def is_shifted_by_one(list1, list2):
    
    if len(list1) != len(list2):
        return False
    return list2[1:] == list1[:-1] and list2[0] != list1[0]

print(is_shifted_by_one([1, 2, 3], [0, 1, 2]))  
print(is_shifted_by_one([1, 2, 3], [3, 1, 2]))  
print(is_shifted_by_one([5, 6, 7], [0, 5, 6]))  
```

```python
def secret_society_name(names):
    
    return ''.join(sorted(n[0].upper() for n in names if n))

friends = ["Alice", "Bob", "Charlie", "Diana"]
print(secret_society_name(friends))  
```

```python
def is_isogram(word):
   
    clean = word.lower()
    return len(set(clean)) == len(clean)

print(is_isogram("Machine"))     # True
print(is_isogram("Programming")) # False
print(is_isogram("Python"))      # True
```

```python
def isogram(sentence):
    words = sentence.lower()
    return set(words) 

print(is_isogram("Machine"))     # True
print(is_isogram("Programming")) # False
print(is_isogram("Python"))    
```

```python
def is_ordered(s):
    
    return list(s) == sorted(s)

print(is_ordered("abc"))     # True
print(is_ordered("aeg"))     # True
print(is_ordered("cba"))     # False
print(is_ordered("loop"))    # False
```

```python
def is_symmetrical(num):
   
    num_str = str(num)
    return num_str == num_str[::-1]

print(is_symmetrical(12321)) 
```
