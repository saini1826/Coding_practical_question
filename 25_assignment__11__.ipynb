```python
def remove_ith_char(string, i):
    if i < 0 or i >= len(string):
        return "Invalid index"
    return string[:i] + string[i+1:]

# Example usage
input_string = input("Enter a string: ")
index = int(input("Enter the index of character to remove: "))

result = remove_ith_char(input_string, index)
print("String after removal:", result)
```

```python
def is_binary_string(s):
    for item in s:
        if item not in ('0', '1'):
            return False
    return True

# Example usage
test_string = input("Enter a string: ")
if is_binary_string(test_string):
    print("The string is a binary string.")
else:
    print("The string is NOT a binary string.")
```

```python
def find_uncommon_words(str1, str2):
    words1 = set(str1.split())
    words2 = set(str2.split())

    uncommon = words1.symmetric_difference(words2)

    return list(uncommon)

# Example usage
string1 = input("Enter first string: ")
string2 = input("Enter second string: ")

result = find_uncommon_words(string1, string2)
print("Uncommon words:", result)
```

```python
def find_duplicates(s):
    s = s.lower()  # Optional: make case-insensitive
    char_count = {}
    duplicates = set()

    for char in s:
        if char != ' ':  # Ignore spaces
            if char in char_count:
                duplicates.add(char)
            else:
                char_count[char] = 1

    return list(duplicates)

# Example usage
input_str = input("Enter a string: ")
result = find_duplicates(input_str)

if result:
    print("Duplicate characters:", result)
else:
    print("No duplicate characters found.")
```

```python
import string

def contains_special_char(s):
    special_chars = string.punctuation
    for char in s:
        if char in special_chars:
            return True
    return False

# Example usage
user_input = input("Enter a string: ")
if contains_special_char(user_input):
    print("The string contains special characters.")
else:
    print("The string does NOT contain any special characters.")
```
