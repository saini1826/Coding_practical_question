```python
def extract_unique_values(d):
    value_count = {}
    for value in d.values():
        value_count[value] = value_count.get(value, 0) + 1

    unique_values = [val for val, count in value_count.items() if count == 1]
    return unique_values

# Example usage
sample_dict = {
    'a': 10,
    'b': 20,
    'c': 10,
    'd': 30,
    'e': 40,
    'f': 20
}

result = extract_unique_values(sample_dict)
print("Unique values:", result)
```

```python
def sum_dictionary_values(d):
    return sum(d.values())

# Example usage
sample_dict = {
    'a': 10,
    'b': 20,
    'c': 30,
    'd': 40
}

total = sum_dictionary_values(sample_dict)
print("Sum of all values:", total)
```

```python
def merge_dictionaries(dict1, dict2):
    merged = {**dict1, **dict2}
    return merged

# Example usage
dict_a = {'a': 1, 'b': 2}
dict_b = {'b': 3, 'c': 4}

merged_dict = merge_dictionaries(dict_a, dict_b)
print("Merged dictionary:", merged_dict)
```

```python
def convert_to_float_dict(pairs):
    float_dict = {}
    for key, value in pairs:
            float_dict[key] = float(value)
    
    return float_dict

# Example usage
key_value_list = [('a', '10'), ('b', '20.5'), ('c', '30')]
result = convert_to_float_dict(key_value_list)
print("Float dictionary:", result)
```

```python
def flat_list(dict1):
    flat = {}
    for item in dict1:
        key, value = item
        flat[key] = value

    return flat

kv_list = [('name', 'Alice'), ('age', 30), ('city', 'Delhi')]
flat_dict = flat_list(kv_list)
print("Flat Dictionary:", flat_dict)
```

```python
def sorted_dict(data):
    reverse = sorted(data.items(), key=lambda x:x[0])
    return reverse

data = {'banana': 3, 'apple': 5, 'cherry': 2, 'date': 4}
result = sorted_dict(data)
print(result)
```
